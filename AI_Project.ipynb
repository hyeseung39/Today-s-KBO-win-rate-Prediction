```python
# 데이터 파일이 너무 길고 방대하여 GPT여도 불가능 하기 때문에 키워드가 포함된 파일만 읽어와서 답을 할 수 있는지 시험해봄.
import os
import pandas as pd
import openai
from openai import OpenAI

# Ollama 서버의 API 엔드포인트 및 API 키 설정
client = OpenAI(
    base_url='http://localhost:11434/v1',  # Ollama의 로컬 서버 주소
    api_key='ollama',  # 필수이지만 실제로 사용되지는 않음
)

# 특정 키워드와 관련된 CSV 파일을 읽어 데이터를 로드하는 함수
def load_relevant_data(directory_path, keyword):
    data_content = ""
    file_names = []  # 파일 이름을 저장할 리스트
    for filename in os.listdir(directory_path):
        if keyword in filename:  # 파일 이름에 키워드가 포함된 경우
            file_path = os.path.join(directory_path, filename)
            if os.path.isfile(file_path) and filename.endswith(".csv"):  # CSV 파일만 읽음
                file_names.append(filename)  # 파일 이름 저장
                
                # CSV 파일을 pandas DataFrame으로 읽기
                df = pd.read_csv(file_path)
                
                # DataFrame을 문자열로 변환
                data_content += df.to_string(index=False) + "\n"
                
    return data_content, file_names

# 디렉토리에서 특정 키워드와 관련된 데이터와 파일 이름 로드
directory_path = './Pitchers relative record'  # 데이터를 저장한 디렉토리 경로
question = "Which hitter is Elias stronger against?"  # 예시 질문

# 키워드를 추출하여 관련 파일만 로드
keyword = "Elias"  # 질문에서 키워드를 추출하거나 지정
context_data, file_names = load_relevant_data(directory_path, keyword)

# 사용자의 질문에 데이터를 참고하여 답변 생성
def ask_question_with_context(question, context_data, file_names):
    # 파일 이름을 컨텍스트에 추가
    file_names_str = ", ".join(file_names)
    response = client.chat.completions.create(
        model="phi3.5",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"The following files were loaded: {file_names_str}. The following context is provided for reference: {context_data}, {question}",}
        ]
    )
    return response.choices[0].message.content  # 응답의 내용을 반환

# 질문에 대한 답변 생성
answer = ask_question_with_context(question, context_data, file_names)

print(f"Question: {question}")
print(f"Answer: {answer}")
```

```python
# 파일을 json으로 변환하기 위해서 시도해봄, 첫번째 시도에서 한글을 생각안하고 코드를 작성하여 json파일이 깨짐, 이를 고려하여 다시 코드 작성
import os
import pandas as pd
import json

def convert_csv_to_json(csv_file_path, json_file_path):
    # CSV 파일을 읽을 때 인코딩을 'utf-8'로 지정
    try:
        df = pd.read_csv(csv_file_path, encoding='utf-8')
    except UnicodeDecodeError:
        print(f"UnicodeDecodeError: 문제 발생 - {csv_file_path}. UTF-8로 인코딩하여 다시 시도합니다.")
        df = pd.read_csv(csv_file_path, encoding='euc-kr')  # 또는 다른 인코딩 시도
    
    # DataFrame을 JSON으로 변환
    data = df.to_dict(orient='records')
    
    # JSON 파일로 저장할 때도 인코딩을 'utf-8'로 지정
    with open(json_file_path, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)  # ensure_ascii=False로 설정하여 한글이 제대로 저장되게 함

    print(f"CSV 파일 {csv_file_path}이(가) JSON 파일 {json_file_path}로 변환되었습니다.")

def convert_all_csv_in_directory(directory_path):
    for filename in os.listdir(directory_path):
        if filename.endswith('.csv'):
            csv_file_path = os.path.join(directory_path, filename)
            json_file_path = os.path.join(directory_path, filename.replace('.csv', '.json'))
            convert_csv_to_json(csv_file_path, json_file_path)

# CSV 파일들이 있는 디렉토리 경로
directory_path = './Pitchers relative record'
convert_all_csv_in_directory(directory_path)
```
